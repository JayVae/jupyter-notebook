```python
import pandas as pd
import numpy as np
df2=pd.read_excel(r'C:\Users\dell\Desktop\BX\20170802\冰箱数据整合.xlsx',sheetname='Sheet1')
df2.dtypes
```

```python
a=df2['年'].values.astype(str)
b=df2['月'].values.astype(str)
```

```python
a
```

```python
df2['week2']=pd.Series(a.tolist()).str.cat(pd.Series(b.tolist()),sep='-')
df2.head(20)
```

```python
df2white=df2[df2['颜色']=='白色']
```

```python
df2white2=pd.Series(df2white.groupby('week2')['PC'].sum())
```

```python
from datetime import datetime
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
%matplotlib inline
xs = [datetime.strptime(d, '%Y-%m').date() for d in df2white2.index]
ys = df2white2.values
# 配置横坐标
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
# Plot
plt.plot(xs, ys)
```
