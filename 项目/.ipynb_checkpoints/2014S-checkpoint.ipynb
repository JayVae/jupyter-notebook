```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel(r'C:\Users\dell\Desktop\BX\20170802\冰箱数据整合.xlsx',sheetname='Sheet1')
df.dtypes
```

```python
df2015=df[df['年']==2014]
```

```python
len(df2015['VIB'].unique())
```

```python
df2015['PC'].sum()
```

```python
pc2015=pd.Series(df2015.groupby('VIB')['PC'].sum())
pc2015.describe()
```

```python
pc2015.sort_values(ascending=False).head(20)
```

```python
plt.rc('font', family='SimHei', size=13)
pc2015.sort_values(ascending=False).head(20).plot(kind='bar',figsize=(15,8))
plt.ylabel('销量',fontsize=13)
plt.xlabel('型号',fontsize=13)
plt.title('2014销量-型号',fontsize=23)
plt.tick_params(labelsize=12,axis='both')
```

```python
price2015=pd.Series(df2015.groupby('供价')['PC'].sum())
pr2015=price2015[price2015>0]
list=[]
for key in pr2015.index:
     i=1
     while i <=pr2015[key]:
         list.append(key)
         i=i+1
pr=pd.Series(list)
plt.figure(figsize=(20,8))
sns.distplot(pr)
plt.xlabel('供价',fontsize=20)
plt.title('2014供价-销量分布图',fontsize=25)
plt.tick_params(labelsize=14,axis='both')
```

```python
zongqj2015=pd.DataFrame(df2015.groupby(u'总容积')['PC'].sum())
zongqj2015['容积区间']='小于200'
for key in zongqj2015.index:
    if (key<200):
        zongqj2015.loc[key,'容积区间']='小于200'
    elif (key<300):
        zongqj2015.loc[key,'容积区间']='大于200小于300'
    else:
        zongqj2015.loc[key,'容积区间']='大于300'

zongqj22015=pd.Series(zongqj2015.groupby(u'容积区间')['PC'].sum())
plt.rc('font', family='SimHei', size=10)
zongqj22015.plot(kind='pie',autopct='%.2f',figsize=(8,8))
plt.title('2014总容积区间',fontsize=10)
```

```python
df2015['llrate']=round(df2015[df2015['冷冻室容积']>0]['冷冻室容积'].astype('float')/df2015[df2015['冷藏室容积']>0]['冷藏室容积'],1)
llrate=pd.Series(df2015.groupby(u'llrate')['PC'].sum())
llrate.plot(kind='bar',figsize=(7,4))
plt.ylabel('销量',fontsize=18)
plt.xlabel('冷冻-冷藏容积比',fontsize=18)
plt.title('2014冷冻-冷藏容积比',fontsize=20)
plt.tick_params(labelsize=12,axis='both')
```
