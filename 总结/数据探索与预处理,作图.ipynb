### 7.4 多个图 subplot

```python
import matplotlib.pyplot as plt
import numpy as np

fig, big_axes = plt.subplots(figsize=(10, 15.0) , nrows=3, ncols=1, sharey=True) 

for row, big_ax in enumerate(big_axes, start=1):
    big_ax.set_title("Subplot row %s \n" % row, fontsize=16)

    # Turn off axis lines and ticks of the big subplot 
    # obs alpha is 0 in RGBA string!
    big_ax.tick_params(labelcolor=(0,0,0,0), top='off', bottom='off', left='off', right='off')
    # removes the white frame
    big_ax._frameon = False

delta=2.0
x=np.linspace(-10,10,100)
y=np.sinc(x-delta)    

for i in range(1,10):
    ax = fig.add_subplot(3,3,i)
    plt.plot(x,y*i)
    ax.set_title('Plot title ' + str(i))


fig.set_facecolor('w')
plt.tight_layout()
plt.show()    
```

```python
import pandas as pd
import numpy as np

df=pd.read_excel(r'E:/项目/电力大数据/yh_new.xlsx')
df=df.fillna(0)## 7.

plt.rc('font', family='SimHei', size=20)
plt.rcParams['axes.unicode_minus']=False
loc_list=df['地理区域'].unique()
name=['2015上','2015下','2016上','2016下','2017上']
x=np.arange(len(name))
y=[0,0,0,0,0]
plt.figure()
fig=plt.figure(figsize=(13,8))
style=['ys-','m*-','rD-','bo-','c+--','gv--','kp--','kv-.','kh-.','bx-.']

for i in range(len(loc_list)):
    y[0] = df.loc[df['地理区域'] == loc_list[i]]['2015上用电偏差量'].mean()
    y[1] = df.loc[df['地理区域'] == loc_list[i]]['2015下用电偏差量'].mean()
    y[2] = df.loc[df['地理区域'] == loc_list[i]]['2016上用电偏差量'].mean()
    y[3] = df.loc[df['地理区域'] == loc_list[i]]['2016下用电偏差量'].mean()
    y[4] = df.loc[df['地理区域'] == loc_list[i]]['2017上用电偏差量'].mean()
    plt.plot(x,y,style[i],label=loc_list[i])
    plt.xlim(-0.5,5.5)
    plt.xticks(x,name,rotation=45)

plt.legend(loc='upper right')
plt.title('不同区域平均偏差电量')
plt.savefig(r'E:/项目/电力大数据/不同区域平均偏差电量情况.png')
```
